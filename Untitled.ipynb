```python
import pandas as pd
low_memory=False
%matplotlib inline
```

```python
filename = 'MICRODADOS_ENADE_2017.csv'
df = pd.read_csv(filename,
                 delimiter=';',
                 encoding='iso-8859-1',
                 usecols=['NT_GER','QE_I17', 'CO_RS_I1', 'CO_RS_I2'])
```

```python
df.rename(columns={'NT_GER': 'NOTABRUTADAPROVA',
                   'QE_I17': 'TIPODEESCOLA',
                   'CO_RS_I1': 'DIFICULDADEFORMACAOGERAL',
                   'CO_RS_I2': 'DIFICULDADECOMPONENTEESPECÍFICO'
                  }, inplace=True)
```

```python
df.sample(10).sort_values(by='NOTABRUTADAPROVA', ascending=False)
```

```python
df.query('NOTABRUTADAPROVA > "50.0"')['TIPODEESCOLA'].value_counts().head().plot(kind='barh')
```

```python
df.query('NOTABRUTADAPROVA > "50.0"')['TIPODEESCOLA'].value_counts().head()
```

```python
df.query('NOTABRUTADAPROVA > "50.0" and DIFICULDADECOMPONENTEESPECÍFICO != "*" and DIFICULDADECOMPONENTEESPECÍFICO != "."')[['TIPODEESCOLA', 'NOTABRUTADAPROVA', 'DIFICULDADEFORMACAOGERAL','DIFICULDADECOMPONENTEESPECÍFICO']].sort_values(by='NOTABRUTADAPROVA', ascending=False).head(60)
```
